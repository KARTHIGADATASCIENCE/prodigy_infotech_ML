```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
hd = pd.read_csv('USA Housing Dataset.csv')
```

```python
x = hd[['bedrooms', 'bathrooms', 'sqft_living']]
y = hd['price']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.45, random_state=110)
```

```python
lr = LinearRegression()
lr.fit(x_train, y_train)
```

```python
predict = lr.predict(x_test)
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(y_test, predict, alpha=0.7, color='blue')
plt.xlabel('True Prices')
plt.ylabel('Predicted Prices')
plt.title('True Prices vs Predicted Prices')
plt.plot([y.min(), y.max()], [y.min(), y.max()], color='red', linestyle='--')  # Diagonal line
plt.grid(True)
plt.show()
```

```python
mse = mean_squared_error(y_test, predict)
r2 = r2_score(y_test, predict)
print(f"Mean Squared Error: {mse:.2f}")
print(f"R^2 Score: {r2:.2f}")
```
